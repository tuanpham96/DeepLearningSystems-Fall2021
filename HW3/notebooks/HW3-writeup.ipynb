For each experiment where you vary hyperparameters use **at least three different values**.

### a) Vary the depth of the network and compare results to the baseline

Here I interpret *depth* as `n_layer` variations: [2, 4, 6, 8] where baseline model has 4 layers. To my surprise, more layers actually **degrade** performance (increase loss by 20-30% and decrease accuracy by around 3% per every 2 layers increase), and of course also increase training time. I really thought that increasing depth would allow the model to learn more effectively. Maybe the models need to be trained for longer and increase `dropout_rate` (though the original paper also has `dropout_rate=0.1`) or apply further regularization to avoid overfitting. One thing I'm not sure of is whether the label smoothing step mentioned in the paper is actually implemented here. That might make a difference.

Additionally, from the translation below for (input-2) and the baseline translation of (input-3), I wonder if there's some inherent gender bias in the model. 

````python
#@title Results
display_df = es_final_df.query('exp_name == "baseline" | exp_name == "vary_depth"')\
    .drop(es_cols2drop, axis=1).reset_index(drop=True)
prm_name = display_df.param_name.tolist()[-1]
display_df.loc[display_df.exp_name == 'baseline','param_name'] = prm_name
display_df.loc[display_df.exp_name == 'baseline','param_val'] = es_base_fdf[prm_name].tolist()[0]
display_df.sort_values(by='param_val').reset_index(drop=True)
````

### b) Vary the model dimension (embedding vector length) and compare it to the baseline

The embedding dimension `d_model` is varied [128, 256, 512, 1024] where 512 is the value of the baseline model. The table below shows only the final results, which indicate that increased embedding dimension do not improve performance and excessive increase (compare baseline and `d_model=1024`) could end up harming performance significantly, while training time only differ by a few percents. However, what is not shown in this table is that, during the first 3-4 epochs, higher `d_model` actually benefits performance quite clearly (for both loss and accuracy). This again my point towards the need for higher regularization. 


````python
#@title Results
display_df = es_final_df.query('exp_name == "baseline" | exp_name == "vary_embeddim"')\
    .drop(es_cols2drop, axis=1).reset_index(drop=True)
prm_name = display_df.param_name.tolist()[-1]
display_df.loc[display_df.exp_name == 'baseline','param_name'] = prm_name
display_df.loc[display_df.exp_name == 'baseline','param_val'] = es_base_fdf[prm_name].tolist()[0]
display_df.sort_values(by='param_val').reset_index(drop=True)
````

## 3.3. compare ReLU against GELU & Swish
Compare the baseline trained with ReLU to that using GELU as described in this paper ([arXiv:1606.08415v4](https://arxiv.org/pdf/1606.08415v4.pdf)) & Swish described in this paper ([arXiv:1710.05941v1](https://arxiv.org/pdf/1710.05941v1.pdf)). Note that GELU (Gaussian Error Linear Units) are what GPT-2 and GPT-J are using instead of ReLU.   See you get better convergence or better loss during training and if any of the translations in your tests have changed. Both GELU and Swish are provided by Tensorflow.

---

The results below show that these activation functions do enhance performance, `gelu` gaining about 0.12% compared to `relu` and `swish` about 0.34% in accuracy. The losses also decrease, by 1.29% for `gelu` and 3.34% for `swish` compared to `relu`. The training time differences are not that different, around 1 hour for all three. 

As for translations, for (input-0,1), `gelu` and `relu` are similar, but `gelu` seems to make more sense than `relu` for (input-2) based on the Google translation of the model's output. (input-3) somehow gets turned to a question by `gelu`. (input-4) differs quite a lot with `gelu` generating something that's not really related, while `relu` still retains the word "interesting". `gelu` at least retains the word "learning" for (5). The (input-6,7) are too hard for both but `gelu` translations, at least via the Google translated version of the model, make more sense than `relu`.

For `swish`, it is interesting that (input-0) is much shorter. (input-1) is similar to that for `relu`.  (input-2) also differs but matches the first 4 words when compared with the Google translation. (input-3,4,6,7) are basically garbage translation, but (input-5) interesting matches really well, missing only 1-2 words.

## 3.5. Summary

Training transformer is hard. Not based on my understanding but based on the results with the experiments shown here:
- Increased model depth or dropout rate has the most noticeable degrading effect on the model's performance. 
- Increased embedding dimension only benefits initially for training but eventually decreases performance.
- Increased number of heads or changing the simple `relu` to `gelu` or `swish` only results in really tiny boost in performance and doesn't seem to affect the training time. 
- Varying the number units in FFN layer or inclusion of the positional encoding layer does not seem to have any noticeable effect. 
- Having a smaller vocabulary size interestingly gains noticable performance gain.
